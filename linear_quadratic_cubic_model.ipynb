```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Simple Playing with Original Data:

```python

data = pd.read_csv('trainRegression.csv')
plt.plot(data['X'],data['R'],'*')
X_data = data['X']
Y_data = data['R']


```

# Implemeting a LINEAR MODEL:

```python

matrix1_for_linear = [[X_data[0], sum(X_data)], [sum(X_data), sum(X_data*X_data)]]

matrix2_for_linear = [[sum(Y_data)], [sum(Y_data*X_data)]]
matrix2_for_linear = np.array(matrix2_for_linear)

A_linear = np.linalg.inv(matrix1_for_linear)
B_linear = matrix2_for_linear

print(A_linear) 
print(B_linear) 

theta_1_linear = A_linear.dot(B_linear)[0]
theta_2_linear = A_linear.dot(B_linear)[1]

print('\n\nTheta 1 linear : ', theta_1_linear)
print('Theta 2 linear :', theta_2_linear)

predit_linear = []

for i in X_data:
    v = theta_1_linear + (theta_2_linear*i)
    predit_linear.append(v[0])


plt.plot(X_data.to_numpy(),np.array(predit_linear),'*')


```

# Implementing a QUADRATIC MODEL:

```python

matrix1_for_quadratic = [
    [X_data[0], sum(X_data), sum(X_data*X_data)],
    [sum(X_data), sum(X_data*X_data), sum(X_data*X_data*X_data)],
    [sum(X_data*X_data), sum(X_data*X_data*X_data), sum(X_data*X_data*X_data*X_data)]
    ]

matrix2_for_quadratic = [[sum(Y_data)], [sum(Y_data*X_data)], [sum(Y_data*(X_data*X_data))]]
matrix2_for_quadratic = np.array(matrix2_for_quadratic)

A = np.linalg.inv(matrix1_for_quadratic)
B = matrix2_for_quadratic

print(A) 
print(B)

theta_1_quadratic = A.dot(B)[0]
theta_2_quadratic = A.dot(B)[1]
theta_3_quadratic = A.dot(B)[2]
print('\n\nTheta 1 : ', theta_1_quadratic)
print('Theta 2 : ', theta_2_quadratic)
print('Theta 3 : ', theta_3_quadratic)

predit_quadratic= []

for i in X_data:
    v = theta_1_quadratic + (theta_2_quadratic*i) + (theta_3_quadratic*(i*i))
    predit_quadratic.append(v[0])


plt.plot(X_data.to_numpy(),np.array(predit_quadratic),'*')
```

# Implementing a CUBIC MODEL:

```python
matrix1_for_cubic = [
    [X_data[0], sum(X_data), sum(X_data*X_data), sum(X_data*X_data*X_data)],
    [sum(X_data), sum(X_data*X_data), sum(X_data*X_data*X_data), sum(X_data*X_data*X_data*X_data)],
    [sum(X_data*X_data), sum(X_data*X_data*X_data), sum(X_data*X_data*X_data*X_data), sum(X_data*X_data*X_data*X_data*X_data)],
    [sum(X_data*X_data*X_data), sum(X_data*X_data*X_data*X_data), sum(X_data*X_data*X_data*X_data*X_data), sum(X_data*X_data*X_data*X_data*X_data*X_data)]
    ]

matrix2_for_cubic = [[sum(Y_data)], [sum(Y_data*X_data)], [sum(Y_data*(X_data*X_data))], [sum(Y_data*(X_data*X_data*X_data))]]
matrix2_for_cubic = np.array(matrix2_for_cubic)

A = np.linalg.inv(matrix1_for_cubic)
B = matrix2_for_cubic

print(A) 
print(B)

theta_1_cubic = A.dot(B)[0]
theta_2_cubic = A.dot(B)[1]
theta_3_cubic = A.dot(B)[2]
theta_4_cubic = A.dot(B)[3]
print('\n\nTheta 1 : ', theta_1_cubic)
print('Theta 2 : ', theta_2_cubic)
print('Theta 3 : ', theta_3_cubic)
print('Theta 4 : ', theta_4_cubic)

predit_cubic = []

for i in X_data:
    v = theta_1_cubic + (theta_2_cubic*i) + (theta_3_cubic*(i*i)) + (theta_4_cubic*(i*i*i))
    predit_cubic.append(v[0])



plt.plot(X_data.to_numpy(),np.array(predit_cubic),'*')
```
